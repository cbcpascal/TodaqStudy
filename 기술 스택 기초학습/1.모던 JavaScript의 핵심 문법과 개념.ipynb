## 모던 JavaScript의 핵심 문법과 개념

네, 그렇다면 모던 JavaScript(ES6+)의 주요 특징과 개념들을 설명드리겠습니다. 기존 JavaScript 지식을 기반으로 새로운 개념들을 살펴보겠습니다.

1. **let과 const를 통한 변수 선언**
```javascript
// var의 문제점 (함수 스코프, 호이스팅)
var x = 1;
if (true) {
    var x = 2;  // 같은 x를 참조
}
console.log(x);  // 2

// let과 const (블록 스코프)
let y = 1;
if (true) {
    let y = 2;  // 다른 y
    console.log(y);  // 2
}
console.log(y);  // 1

// const는 재할당 불가
const PI = 3.14;
// PI = 3.15;  // Error!

// 객체나 배열의 경우 내부 값은 변경 가능
const user = { name: 'John' };
user.name = 'Jane';  // 가능
// user = { name: 'Jane' };  // Error!
```

2. **Arrow Functions**
```javascript
// 기존 함수 표현
function add(a, b) {
    return a + b;
}

// 화살표 함수
const add = (a, b) => a + b;

// this 바인딩의 차이
class Person {
    constructor(name) {
        this.name = name;
    }
    
    // 일반 함수: this가 호출 컨텍스트에 바인딩
    sayHiLater() {
        setTimeout(function() {
            console.log(`Hi, ${this.name}`);  // this.name은 undefined
        }, 1000);
    }
    
    // 화살표 함수: 렉시컬 스코프의 this를 유지
    sayHiLaterArrow() {
        setTimeout(() => {
            console.log(`Hi, ${this.name}`);  // 정상 작동
        }, 1000);
    }
}
```

3. **구조 분해 할당 (Destructuring)**
```javascript
// 객체 구조 분해
const person = {
    name: 'John',
    age: 30,
    address: {
        city: 'Seoul',
        country: 'Korea'
    }
};

const { name, age, address: { city } } = person;
console.log(name, age, city);  // John 30 Seoul

// 배열 구조 분해
const numbers = [1, 2, 3, 4, 5];
const [first, second, ...rest] = numbers;
console.log(first, second, rest);  // 1 2 [3, 4, 5]
```

4. **Spread/Rest 연산자**
```javascript
// Spread: 배열이나 객체를 펼치기
const arr1 = [1, 2, 3];
const arr2 = [...arr1, 4, 5];  // [1, 2, 3, 4, 5]

const obj1 = { foo: 'bar', x: 42 };
const obj2 = { ...obj1, y: 13 };  // { foo: 'bar', x: 42, y: 13 }

// Rest: 나머지 매개변수
const sum = (...numbers) => {
    return numbers.reduce((acc, cur) => acc + cur, 0);
};
console.log(sum(1, 2, 3, 4));  // 10
```

5. **Promise와 async/await**
```javascript
// Promise 기반 비동기 처리
const fetchUser = () => {
    return new Promise((resolve, reject) => {
        setTimeout(() => {
            const user = { id: 1, name: 'John' };
            resolve(user);
            // 에러 발생 시: reject(new Error('User not found'));
        }, 1000);
    });
};

// Promise 사용
fetchUser()
    .then(user => console.log(user))
    .catch(error => console.error(error));

// async/await 사용 (더 깔끔한 비동기 처리)
async function getUser() {
    try {
        const user = await fetchUser();
        console.log(user);
    } catch (error) {
        console.error(error);
    }
}
```

6. **템플릿 리터럴**
```javascript
const name = 'John';
const age = 30;

// 기존 방식
console.log('My name is ' + name + ' and I am ' + age + ' years old');

// 템플릿 리터럴
console.log(`My name is ${name} and I am ${age} years old`);

// 여러 줄 문자열
const html = `
    <div>
        <h1>${name}</h1>
        <p>${age}</p>
    </div>
`;
```

# 모던 자바스크립트에서 `var`의 문제점

## 1. 함수 스코프(Function Scope)만 지원
- `var`는 함수 내에서만 유효한 함수 스코프를 가짐
- 블록 스코프(Block Scope)를 무시하므로 예상치 못한 동작 발생 가능

```javascript
if (true) {
  var x = 10;
}
console.log(x); // 10 - 블록을 벗어났는데도 접근 가능 (문제점)

for (var i = 0; i < 5; i++) {
  // loop
}
console.log(i); // 5 - 반복문이 끝났는데도 i에 접근 가능 (문제점)
```

## 2. 호이스팅(Hoisting) 문제
- `var`로 선언된 변수는 스코프의 최상단으로 끌어올려짐(호이스팅)
- 선언 전에 사용 가능하지만 값은 `undefined`가 되어 혼란 야기

```javascript
console.log(a); // undefined (에러가 아니라 문제)
var a = 5;
```

위 코드는 실제로 다음과 같이 동작:
```javascript
var a;          // 선언 호이스팅
console.log(a); // undefined
a = 5;          // 할당
```

## 3. 재선언 가능
- 같은 스코프 내에서 동일한 변수명으로 여러 번 선언 가능
- 실수로 변수를 재정의해도 에러가 발생하지 않음

```javascript
var x = 1;
var x = 2; // 에러 없이 재선언 가능 (문제점)
console.log(x); // 2
```

## 4. 전역 변수 오염
- 함수 외부에서 `var`로 선언하면 전역 객체의 속성이 됨
- 브라우저 환경에서는 `window` 객체에 속성으로 추가됨

```javascript
var globalVar = "I'm global";
console.log(window.globalVar); // "I'm global"
```

## 모던 자바스크립트의 해결책: `let`과 `const`
- `let`과 `const`는 블록 스코프(Block Scope)를 지원
- 호이스팅은 발생하지만 TDZ(Temporal Dead Zone)로 인해 선언 전 사용 불가
- 재선언 불가능
- 전역으로 선언해도 `window` 객체의 속성이 되지 않음

이러한 `var`의 문제점들 때문에 ES6 이후에는 `let`과 `const` 사용을 권장합니다.

# 모던 자바스크립트의 Spread(전개)와 Rest(나머지) 연산자

Spread와 Rest 연산자는 모두 `...`(점 3개) 기호로 표기되지만, 사용되는 컨텍스트에 따라 다르게 동작합니다.

## 1. Spread 연산자 (전개 연산자)

배열, 객체, 문자열 등의 **이터러블(iterable)한 값**을 개별 요소로 "펼쳐서(전개)" 사용할 때 활용합니다.

### 배열에서의 사용 예시

```javascript
// 배열 복사
const arr1 = [1, 2, 3];
const arr2 = [...arr1]; // [1, 2, 3] (새로운 배열 생성)

// 배열 합치기
const arr3 = [0, ...arr1, 4, 5]; // [0, 1, 2, 3, 4, 5]

// 함수 인자로 전달
const numbers = [1, 2, 3];
Math.max(...numbers); // Math.max(1, 2, 3)과 동일
```

### 객체에서의 사용 예시 (ES2018+)

```javascript
// 객체 복사
const obj1 = { a: 1, b: 2 };
const obj2 = { ...obj1 }; // { a: 1, b: 2 } (새로운 객체 생성)

// 객체 합치기
const obj3 = { ...obj1, c: 3, d: 4 }; // { a: 1, b: 2, c: 3, d: 4 }

// 객체 업데이트 (동일 키는 뒤에 오는 값이 우선)
const obj4 = { ...obj1, a: 10 }; // { a: 10, b: 2 }
```

## 2. Rest 연산자 (나머지 연산자)

함수의 매개변수나 배열/객체 구조 분해 할당 시 **나머지 값들을 묶어서** 사용할 때 활용합니다.

### 함수 매개변수에서의 사용

```javascript
function sum(...numbers) {
  return numbers.reduce((acc, cur) => acc + cur, 0);
}

sum(1, 2, 3, 4); // 10 (numbers는 [1, 2, 3, 4] 배열)
```

### 배열 구조 분해 할당에서의 사용

```javascript
const [first, second, ...rest] = [1, 2, 3, 4, 5];
console.log(first); // 1
console.log(second); // 2
console.log(rest); // [3, 4, 5]
```

### 객체 구조 분해 할당에서의 사용 (ES2018+)

```javascript
const { a, b, ...rest } = { a: 1, b: 2, c: 3, d: 4 };
console.log(a); // 1
console.log(b); // 2
console.log(rest); // { c: 3, d: 4 }
```

## 주의사항

1. **얕은 복사(Shallow Copy)**: Spread 연산자는 1레벨 깊이만 복사합니다. 중첩된 객체/배열은 참조가 복사됩니다.
   ```javascript
   const nested = { a: { b: 1 } };
   const copy = { ...nested };
   copy.a.b = 2; // nested.a.b도 2로 변경됨
   ```

2. **객체 Spread는 ES2018부터 표준**: 최신 브라우저/Node.js에서만 지원

3. **Rest는 마지막에만 올 수 있음**
   ```javascript
   const [...rest, last] = [1, 2, 3]; // SyntaxError
   ```

Spread와 Rest 연산자는 모던 자바스크립트에서 불변성(Immutability)을 유지하면서 데이터를 다루는 데 매우 유용하게 사용됩니다.

# Rest 연산자와 `reduce` 메서드 이해하기

Rest 연산자가 함수 매개변수에서 어떻게 작동하는지, 그리고 함께 사용된 `reduce` 메서드까지 자세히 설명드리겠습니다.

## Rest 연산자 기본 개념

```javascript
function sum(...numbers) {
  // numbers는 전달된 모든 인자를 배열로 묶음
  return numbers.reduce((acc, cur) => acc + cur, 0);
}

sum(1, 2, 3, 4); // 결과: 10
```

1. `...numbers`는 **함수에 전달된 모든 인자들을 배열로 묶어줍니다.**
   - `sum(1, 2, 3, 4)` 호출 시 → `numbers = [1, 2, 3, 4]`

## `reduce` 메서드 작동 원리

`reduce`는 배열의 각 요소를 처리해 단일 값으로 만드는 메서드입니다.

```javascript
arr.reduce((누적값, 현재값) => { /* 로직 */ }, 초기값);
```

예제의 `reduce` 동작 과정:

```javascript
[1, 2, 3, 4].reduce((acc, cur) => acc + cur, 0);

// 1회차: acc = 0, cur = 1 → 0 + 1 = 1
// 2회차: acc = 1, cur = 2 → 1 + 2 = 3
// 3회차: acc = 3, cur = 3 → 3 + 3 = 6
// 4회차: acc = 6, cur = 4 → 6 + 4 = 10
// 최종 결과: 10
```

## 단계별 해석

1. **함수 호출**
   ```javascript
   sum(1, 2, 3, 4);
   ```

2. **Rest 연산자 동작**
   ```javascript
   function sum(...numbers) {
     // numbers = [1, 2, 3, 4]로 변환
   }
   ```

3. **reduce 실행**
   ```javascript
   [1, 2, 3, 4].reduce((acc, cur) => acc + cur, 0);
   ```

4. **실행 과정 시각화**

   | 회차 | acc (누적값) | cur (현재값) | 결과  |
   |------|-------------|-------------|-------|
   | 1    | 0 (초기값)   | 1           | 0 + 1 = 1 |
   | 2    | 1           | 2           | 1 + 2 = 3 |
   | 3    | 3           | 3           | 3 + 3 = 6 |
   | 4    | 6           | 4           | 6 + 4 = 10 |

## 다른 예제로 이해하기

### 1. 평균 계산
```javascript
function average(...nums) {
  const sum = nums.reduce((acc, num) => acc + num, 0);
  return sum / nums.length;
}

average(10, 20, 30); // (10+20+30)/3 = 20
```

### 2. 최대값 찾기
```javascript
function findMax(...numbers) {
  return numbers.reduce((max, num) => num > max ? num : max, -Infinity);
}

findMax(5, 2, 9, 1); // 9
```

## Rest vs 일반 매개변수

```javascript
function example(a, b, ...rest) {
  console.log(a);    // 1
  console.log(b);    // 2
  console.log(rest); // [3, 4, 5]
}

example(1, 2, 3, 4, 5);
```

- `a`, `b`는 일반 매개변수로 먼저 할당
- 남은 인자들은 `...rest`에 배열로 할당
